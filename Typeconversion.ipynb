##
What is Type Conversion?
Definition

Type Conversion is the process of converting one data type into another data type.

```python
a = 5
print(type(a))
```

### Types of Type Conversion

There are 2 types:

Implicit Type Conversion
Explicit Type Conversion (Type Casting)
## 1. Implicit Type Conversion
Definition

Python automatically converts one data type into another without user intervention.

Also called Automatic Type Conversion.

```python
a = 40 
a = float(a)
print(type(a))
```

```python
a = 50 + 2.5
print(type(a))
```

```python
print(a)
```

## 2. Explicit Type Conversion (Type Casting)
Definition

When the programmer manually converts one data type into another using built-in functions.

```python
num_int = int(3.14)
print(num_int)
```

```python
num_float = (float(75))
print(num_float)
```

```python
message = str(45)
print(message)
```

```python
char_b = 'B'
a = ord(char_b)
print(a)
```

```python
a= 10
b= 3
floor_division = a//b
print(floor_division)
```

```python
a =4
b=5
modulo = a%b
print(modulo)
```

```python
x = 15
print(x)
```
